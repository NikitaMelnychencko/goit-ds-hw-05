Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

```python

df_author_count = (
    df
    .groupby('author')                                  # групуємо за автором
    .agg(count=('name', 'count'))            # рахуємо кількість назв книг
    .reset_index()
    )
df_author_count
```

```python
df_author_count.max()["author"]
```

Скільки книг цього автора? Відповідь:

Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
df_author_count_mid = (
    df
    .groupby('author')
    .agg(mid_rating=('user_rating', lambda x: (x.max() + x.min()) / 2))
    .reset_index()
)
df_author_count_mid
```

У якого автора середній рейтинг мінімальний?

```python
df_author_count_mid.min()["author"]
```

Який у цього автора середній рейтинг?

```python
df_author_count_mid.min()["mid_rating"]
```

З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
df_author_count

result_frame = pd.concat(
    [df_author_count_mid, df_author_count["count"]],
    axis=1
)
result_frame
```

Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

```python
result_frame = result_frame.sort_values(by=["count", "mid_rating"], ascending=[False, False] )
result_frame
```

```python

# 1. Вкажіть автора, якого хочете проаналізувати
author_name = "Jeff Kinney"

# 2. Фільтрація датафрейму за цим автором
df_author = df[df['author'] == author_name]

# 3. Групування за роком та підсумовування продажів
#    Припустимо, що в колонці 'sold' — кількість проданих книг
sales_per_year = (
    df_author
    .groupby('year')["reviews"]
    .sum()
    .sort_index()
)
print(sales_per_year)

# 4. Побудова графіка
plt.figure(figsize=(10, 5))
sales_per_year.plot(kind='bar', color='teal')

plt.title(f"Продажі книг {author_name} по роках", fontsize=14)
plt.xlabel("Рік", fontsize=12)
plt.ylabel("Кількість reviews", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
