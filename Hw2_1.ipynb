```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://uk.wikipedia.org/wiki/Населення_України"
headers = {"User-Agent": "Mozilla/5.0"}
page = requests.get(url, headers=headers).text
df = pd.read_html(
    page,
    encoding="utf-8",
    decimal=',',  # Вказуємо, що десятковий роздільник - кома
    thousands=' ',  # Вказуємо, що тисячі розділяються пробілом
    index_col=0  # Перша колонка буде індексом
)[13]

df
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
new_df = df.replace(to_replace="—", value=pd.NA)
new_df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
new_df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df = new_df.apply(pd.to_numeric, errors='coerce')

print("Типи колонок після конвертації:")
print(df.dtypes)
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
column_names = df.columns

for column in column_names:
  nan_values = df[column].isnull().sum()
  if nan_values > 0:
    print(f"{column}: {nan_values}")
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df[:-1]
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
mid_value = df.mean()

df.fillna(mid_value)
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
year = "2019"
mid_value_by_year = mid_value[year]
column_by_year = df[year]

df[column_by_year > mid_value_by_year][year]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df["2014"].idxmax()
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python

# Виберіть рік та регіони
year = "2019"
regions = df.index

# Отримуємо дані для конкретного року
selected_data = df[year]

# Створюємо простий стовпчиковий графік
plt.figure(figsize=(14, 8))
plt.bar(regions, selected_data, color='skyblue', edgecolor='black')

plt.xlabel('Регіон')
plt.ylabel('Коефіцієнт народжуваності')
plt.title(f'Коефіцієнт народжуваності по регіонах у {year} році')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
